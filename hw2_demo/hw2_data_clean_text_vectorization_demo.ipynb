## create simple data 

```python
raw_docs = ["Here are some very simple basic sentences.",
"They won't be very interesting, I'm afraid.",
"The point of these examples is to _learn how basic text cleaning works_ on *very simple* data."]
```

## import russian news data 

```python
import json
with open('/Users/gongqian/Desktop/spark_test/Q3/russian/train.json') as data_file:    
    data = json.load(data_file)


print(data[0]['text'])

print(data[0]['id'])

print(data[0]['sentiment'])

raw_docs = []
sentiments=[]
sen_dic={"positive":1,"negative":-1,"neutral":0}
for i in range(1000):
    raw_docs.append(data[i]['text'])
    sentiments.append(sen_dic[data[i]['sentiment']])

print repr(raw_docs[15]).decode("unicode-escape")
print sentiments[15]
```

## tokenize text into words 

```python
# Tokenizing text into bags of words
import nltk
from nltk.tokenize import word_tokenize
tokenized_docs = [word_tokenize(doc) for doc in raw_docs]

print repr(tokenized_docs[2]).decode("unicode-escape")
```

## remove punctuation 

```python
# Removing punctuation
import re
import string
regex = re.compile('[%s]' % re.escape(string.punctuation)) #see documentation here: http://docs.python.org/2/library/string.html

tokenized_docs_no_punctuation = []

for review in tokenized_docs:
    new_review = []
    for token in review:
        new_token = regex.sub(u'', token)
        if not new_token == u'':
            new_review.append(new_token)
    
    tokenized_docs_no_punctuation.append(new_review)
    

print repr(tokenized_docs_no_punctuation[2]).decode("unicode-escape")
```

## clean text of stopwords 

```python
# Cleaning text of stopwords
from nltk.corpus import stopwords

tokenized_docs_no_stopwords = []

for doc in tokenized_docs_no_punctuation:
    new_term_vector = []
    for word in doc:
        if not word in stopwords.words('russian'):
            new_term_vector.append(word)
    
    tokenized_docs_no_stopwords.append(new_term_vector)

print repr(tokenized_docs_no_stopwords[2]).decode("unicode-escape")
```

## word stemming 

```python
# Stemming and Lemmatizing
from nltk.stem.porter import PorterStemmer
from nltk.stem.snowball import SnowballStemmer
from nltk.stem.wordnet import WordNetLemmatizer

porter = PorterStemmer()
snowball = SnowballStemmer('russian')
wordnet = WordNetLemmatizer()

preprocessed_docs = []

for doc in tokenized_docs_no_stopwords:
    final_doc = []
    for word in doc:
        final_doc.append(porter.stem(word))
        #final_doc.append(snowball.stem(word))
        #final_doc.append(wordnet.lemmatize(word))
    
    preprocessed_docs.append(final_doc)

print repr(preprocessed_docs[2]).decode("unicode-escape")
```

## create svm format data feature vector 

```python
pre_svm=[]
for i in range(3):
    string = ' '.join(preprocessed_docs[i])
    pre_svm.append(string)
```

```python
print repr(pre_svm[2]).decode("unicode-escape")
```
